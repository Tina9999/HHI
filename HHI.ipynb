```python
import pandas as pd 
from IPython.display import display, HTML 
import matplotlib.pyplot as plt 
import seaborn as sns 

# Putanja do Excel fajla 
file_path = r'C:\Users\Korisnik\Desktop\HHI_index.xlsx' # Putanja do vaseg Excel fajla 

# Ucitavanje podataka sa specifinog sheet-a 
df = pd.read_excel(file_path, sheet_name= 'Sheet1') # Zameni 'HHI_racun' sa starnim nazivom sheet-a 

# Pretvaranje kolona u numericke vrednosti 
#df[ 'Total '] = pd. to_numeric(df[ 'Total '], errors= 'coerce'). fillna(0) 
df['Bads'] = pd. to_numeric(df[ 'Bads'], errors='coerce'). fillna (0) 
df ['Goods'] = pd. to_numeric(df['Goods'], errors= 'coerce'). fillna(0) 
df['Total'] = (df['Goods'] + df['Bads']) 

# Racunanje ukupnog Total-a (suma svih vrednosti u koloni 'Total ') 
ukupno_total = df['Total'].sum ( ) # Ukupna suma svih Total vrednosti 

# Racunanje DR (Total / Bads) i pretvaranje u procente 
df ['DR'] = (df['Bads'] / df['Total']).fillna(0) #* 100 # Multiplikacija sa 100 da bi bio u procentima

#Racunanje Ucesce (Total po bucketima / Ukupno Total) i pretvaranje u procente 
df['Ucesce'] = (df['Total'] / ukupno_total) #* 100 # Ucesce u procentima 

# Racunanje Ucesce^2 i pretvaranje u procente 
df['Ucesce^2'] = (df['Ucesce'] ** 2) # Ucesce^2 je vec u procentima 

# Izracunavanje HHI indeksa (suma kolone Ucesce ^2) 
hhi_index = df['Ucesce^2'].sum() 
hhi_index_percentage = hhi_index * 100 # Pretvoreno u procente 

# Snimanje rezultata u Excel fajl (dodajemo nove kolone) 
with pd.ExcelWriter(file_path, engine='openpyxl', mode='a', if_sheet_exists='replace') as writer: 
     df.to_excel(writer, sheet_name= 'HHI', index=False) 
    #df.to_excel (file_path, sheet_name='HHI_racun', index=False) 

over_20_percent = (df['Ucesce']> 0.20).sum() 
#print(f"Broj bucket-a sa ucescem vecim od 20%: (over_20_percent}") 

def highlight_ucesce (val): 
    color = 'red' if val > 0.20 else 'black' # Ako je ucesce vece od 20%, oboji cruenom 
    return f'color: {color}' 
    
styled_df = df.style.map(highlight_ucesce, subset=[ 'Ucesce']) 

display (styled_df) 

#uslovi za dodelu zute, zelene iLi cruene boje na rezultat Testa distribucije rejting kategorije 
if over_20_percent <= 1:
    color = 'black'
    background_color = 'green'
elif 20 < hhi_index_percentage <= 40:
    color = 'black'
    background_color = 'yellow'
else:
    color = 'black'
    background_color = 'red'

display(HTML(f"<h4><b style = 'color: {color};background-color: {background_color}; border: 1px solid black; padding: 3px;'>Test distribucije rejting kategorije = {over_20_percent:.0f}</b></h4>"))


#uslovi za dodelu zute, zelene iLi cruene boje na rezultat testa HHI indexa
if over_20_percent <= 1:
    color = 'black'
    background_color = 'green'
elif 20 < hhi_index_percentage <= 40:
    color = 'black'
    background_color = 'yellow'
else:
    color = 'black'
    background_color = 'red'

display(HTML(f"<h4><b style = 'color: {color};background-color: {background_color}; border: 1px solid black; padding: 3px;'>HHI Index = {hhi_index_percentage:.2f}%</b></h4>"))












```
